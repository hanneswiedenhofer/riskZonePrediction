```python
import pandas as pd
import numpy as np
import shapely
import datetime
from random import randint
import os
os.environ['PROJ_LIB'] = r''
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from shapely.geometry import Point
import geopy.distance
import time
import matplotlib.patches as mpatches
start = time.time()
```

```python
keep_only_every_nth_row = 5
allarm_limit = 5000
min_points = 20
time_diff_between_routes = datetime.timedelta(days=0, seconds=0, microseconds=0, milliseconds=0, minutes=0, hours=4, weeks=0)
```

```python
df = pd.read_csv("percorsoMezzi.csv")
df["DataOraAcquisizione"] = pd.to_datetime(df["DataOraAcquisizione"])
df = df.iloc[::keep_only_every_nth_row, :]
```

```python
eventi_df = pd.read_csv("my_eventi.csv")
```

```python
allarms = eventi_df[["Lat","Lon"]]
allarm_points = []
```

```python
for index, row in allarms[:allarm_limit].iterrows():
    allarm_points.append(Point(row["Lon"], row["Lat"]))
```

```python
dfs = dict(tuple(df.groupby('IdMezzo')))
```

```python
#my_df = dfs["56fa8952-64ed-e711-80be-0050569357e4"]
```

```python
intervals = {}
for key in dfs.keys():
    dfs[key]["distance"] = dfs[key]["DataOraAcquisizione"].diff()
    interval = dfs[key][dfs[key]["distance"] > time_diff_between_routes]["DataOraAcquisizione"].tolist()
    intervals[key] = [(np.datetime64('2000-01-01T00:00:00'),'black')]
    for el in interval:
        intervals[key].append((np.datetime64(el),'#%06X' % randint(0, 0xFFFFFF)))
```

```python
my_points = []
```

```python
class route():
    
    def __init__(self, mezzo, route):
        self.mezzo = mezzo
        self.route = route
        self.lon = 0
        self.lat = 0
        self.label = ""
        self.points = []
        self.score = 0
    
    def add_point(self, lon, lat, label):
        my_point = Point(lon, lat)
        my_point = my_point.buffer(0.001)
        self.points.append((my_point, lon, lat, label))
    
    def calc_score(self, count_allarms, distance):
        if distance > 0:
            self.score = count_allarms/distance
        else:
            self.score = 0
    
    def __str__(self):
        return(f"Mezzo: {self.mezzo}, Route: {self.route}, Score: {self.score}")
```

```python
my_points = []
```

```python
for key in list(dfs.keys())[:10]:
    #key = "56fa8952-64ed-e711-80be-0050569357e4"
    my_df = dfs[key]
    for i in range(len(my_df["Latitudine"])):
            for j in range(1,len(intervals[key])):
                if (my_df["DataOraAcquisizione"].iloc[i] <= intervals[key][j][0]) & (my_df["DataOraAcquisizione"].iloc[i] > intervals[key][j-1][0]):
                    my_points.append((my_df["Longitudine"].iloc[i], my_df["Latitudine"].iloc[i], f"{my_df['DataOraAcquisizione'].iloc[i]}", intervals[key][j][1], key))
```

```python
output = {}
for v,w,x,y,z in my_points:
    if (z,y) in output:
        output[z,y].append((v,w,x))
    else:
        output[z,y] = [(v,w,x)]
```

```python
my_routes = []
```

```python
for key in output.keys():
    my_route = route(key[0],key[1])
    for el in output[key]:
        my_route.add_point(el[0],el[1],el[2])
        #my_route.add_point()
    my_routes.append(my_route)
```

```python
for el in my_routes:
    buffered_points = [a_tuple[0] for a_tuple in el.points]
    count = 0
    distance = geopy.distance.distance((el.points[0][1],el.points[0][2]), (el.points[-1][1],el.points[-1][2])).km
    for el1 in buffered_points:
        for el2 in allarm_points:
            if el1.intersects(el2):
                count += 1
    el.calc_score(count,distance)
```

```python
for key in list(dfs.keys())[:10]:
#key = "56fa8952-64ed-e711-80be-0050569357e4"
    fig = plt.figure(figsize=(24,24))
    plt.title(str(key))
    m = Basemap(projection="mill", llcrnrlat=35, urcrnrlat=55, llcrnrlon=-10, urcrnrlon=30, resolution="c")
    m.drawcoastlines()
    my_temp_patches = []
    my_patches = []
    for route in my_routes:
        if route.mezzo == key:
            for point in route.points:
                if (len(route.points) > min_points) & (route.score > 0):
                    m.scatter(point[1],point[2],latlon=True, label=point[3],c=route.route)
                    if route.route not in my_temp_patches:
                        my_temp_patches.append(route.route)
                        my_patches.append(mpatches.Patch(color=route.route,label=f"Score: {route.score}"))
    #red_patch = mpatches.Patch(color='red', label='The red data')
    #blue_patch = mpatches.Patch(color='blue', label='The blue data')

    plt.legend(handles=my_patches)
    plt.show()
```

```python
end = time.time()
print(f"Time elapsed: {(end - start)/60} min")
```
